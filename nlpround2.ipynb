```python
# ORIGINAL
def quotes_parser(key, max_pages):
    list_of_quotes=[]
    for page_number in range(max_pages + 1):
        dynamic_url = f"https://clientsfromhell.net/tag/{key}/page/{page_number}"
        html = requests.get(dynamic_url).content
        soup = bs(html, "html.parser")
        list_of_items = soup.find_all('div',{'class':'w-blog-post-content'})
        for quotes in list_of_items:
#             list_of_quotes.append(re.findall('(?<=Client:).*', _))
            list_of_quotes.append(quotes.text)

    return list_of_quotes
```

```python
from nltk.corpus import stopwords
en_stopwords = stopwords.words('english')

def count_words(word_lists):
    dfs = []
    for list_ in word_lists:
        df = pd.DataFrame()
        for word in set(list_):
            df[word] = [list_.count(word)]
        dfs.append(df)
    return pd.concat(dfs, sort = False).reset_index(drop = True)
```

```python
corpus_felon = quotes_parser(felon, 14)
# corpus_Dunces = quotes_parser(Dunces, 81)
# corpus_Deadbeats = quotes_parser(Deadbeats, 58)
# corpus_Racists = quotes_parser(Racists, 6)
# corpus_Homophobes = quotes_parser(Homophobes, 3)
# corpus_Sexist = quotes_parser(Sexist, 6)
# corpus_Frenemies = quotes_parser(Frenemies, 4)
# corpus_Cryptic = quotes_parser(Cryptic, 51)
# corpus_Ingrates = quotes_parser(Ingrates, 24)
# corpus_ChaoticGood = quotes_parser(Chaotic_Good, 39)
```

```python
blob2 = [TextBlob(x).words.lower() for x in corpus_felon]
blobb = [' '.join(i) for i in blob2]
```

```python
df_Felon = pd.DataFrame(
    data = pd.Series(blobb)
)

df_Felon['Type'] = 'Felon'
```

```python
tfidf = TfidfVectorizer()

word_count_vectors = tfidf.fit_transform(
    df_Felon[0].apply(
        lambda x: ' '.join([w for w in x.split() if w not in en_stopwords])
)).todense().tolist()
```

```python
df2 = pd.DataFrame(
    data = word_count_vectors,
    columns = tfidf.get_feature_names(), # neat method to pull in the column names
)

df2.sum().sort_values(ascending = False)
```
